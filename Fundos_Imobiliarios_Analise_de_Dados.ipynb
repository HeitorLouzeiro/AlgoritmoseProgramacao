```python
from datetime import datetime
import pandas as pd 
import yfinance as yf 
import requests 
from bs4 import BeautifulSoup 
import plotly
import openpyxl
```

```python
url = 'https://www.fundsexplorer.com.br/ranking'
```

```python
agent = {"User-Agent":"Mozilla/5.0"}
response = requests.get(url)

soup = BeautifulSoup(response.text, 'lxml')
table = soup.find_all('table')[0]
df = pd.read_html(str(table), decimal='.', thousands=',')[0]
df
```

```python
df.dtypes
```

```python
df.to_csv('fundos_imob.csv') #salvando dados em um arquivo .csv 
df = pd.read_csv('fundos_imob.csv') #lendo arquivo .cvs
df
```

## Verificando os FIIs não descontados

```python
fiis_ndescontados = df[df['P/VPA'] >= 1]
fiis_ndescontados
```

## Verificando os FIIs descontados

```python
fiis_descontados = df[df['P/VPA'] < 100]
fiis_descontados
```

## Verificando os FIIs descontados com a Liquidez Diária de 5000 negociações diárias  

```python
fiis_descontados_ld5000 = fiis_descontados[fiis_descontados['Liquidez Diária']>5000]
fiis_descontados_ld5000
```
